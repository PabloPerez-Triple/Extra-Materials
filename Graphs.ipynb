### 1. Boxplot (Diagrama de Caja y Bigotes)
__<font color=blue>Cómo interpretarlo</font>__
Un boxplot (o diagrama de caja y bigotes) proporciona un resumen gráfico de la distribución de los datos basado en cinco números: mínimo, primer cuartil (Q1), mediana, tercer cuartil (Q3) y máximo. 
Se compone de:

- __La caja__: Representa el rango intercuartílico (IQR) entre Q1 y Q3 (el 50% central de los datos).

- __La línea en la caja__: La mediana (valor medio).

- __Los bigotes__: Indican la variabilidad fuera de los cuartiles superior e inferior, pero se limitan a 1.5 veces el IQR.

- __Valores atípicos__: Puntos fuera de los bigotes, generalmente marcados como puntos.

Ejemplo de interpretación: Si la línea de la mediana está más cerca de la parte inferior de la caja, indica una distribución sesgada a la izquierda. Los valores atípicos pueden indicar anomalías o desviaciones significativas en tus datos.

__<font color=blue>Casos de uso</font>__

- Identificación de valores atípicos.
- Comparación de distribuciones entre categorías.
- Útil en control de calidad para evaluar la variación.

__<font color=blue>Documentación</font>__

[Boxplot_seaborn](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

```python
# Create boxplot
sns.boxplot(data = df, 
            x = 'time', 
            y = 'total_bill', 
            hue = 'smoker',
            palette = 'husl'
            ) 

# palette options deep, muted, bright, pastel, dark, colorblind
```

### 2. Histograma
__<font color=blue>Cómo interpretarlo</font>__
Un histograma muestra la distribución de frecuencia de datos continuos agrupando los datos en intervalos (bins). El eje x representa los rangos de los datos, mientras que el eje y muestra la frecuencia de los datos dentro de cada intervalo.

Ejemplo de interpretación: Si ves una curva en forma de campana, podría sugerir una distribución normal. Un histograma sesgado a la derecha (cola a la derecha) indica una asimetría positiva, y hacia la izquierda una asimetría negativa.

__<font color=blue>Casos de uso</font>__

- Comprender la distribución de una variable.
- Verificar la normalidad de los datos.
- Ideal para conjuntos de datos grandes.

__<font color=blue>Documentación</font>__

[Histogram_seaborn](https://seaborn.pydata.org/generated/seaborn.histplot.html)

```python
# Create histogram
sns.histplot(data=df, 
             x = 'total_bill',
             hue='sex'
            )
```

### 3. Gráfico de dispersión (Scatterplot)
__<font color=blue>Cómo interpretarlo</font>__
Un gráfico de dispersión se utiliza para visualizar la relación entre dos variables continuas. Cada punto representa una observación con valores en los ejes x e y. Ayuda a identificar:

- __Tendencias__: Si existe una relación positiva, negativa o ninguna.

- __Agrupaciones__: Si los puntos tienden a agruparse.

- __Valores atípicos__: Puntos que están lejos de la tendencia general.

Ejemplo de interpretación: Si los puntos siguen una tendencia ascendente, sugiere una correlación positiva (cuando una variable aumenta, la otra también). Una tendencia descendente indica una correlación negativa.

__<font color=blue>Casos de uso</font>__

- Analizar la relación entre dos variables numéricas.
- Detectar tendencias, agrupaciones o valores atípicos en los datos.
- Usado comúnmente en análisis de regresión.

__<font color=blue>Documentación</font>__

[Scatterplot_seaborn](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

```python
# Create scatterplot
sns.scatterplot(data=data, 
                x="total_bill", 
                y="tip",
                hue='smoker')
```

### 4. Gráfico de barras (Barplot)
__<font color=blue>Cómo interpretarlo</font>__
Un gráfico de barras se utiliza para mostrar datos categóricos y comparar valores entre diferentes categorías. La longitud de las barras representa la magnitud de la variable para cada categoría.

Ejemplo de interpretación: Las barras más altas indican valores más grandes, mientras que las más cortas representan valores menores. Si todas las barras tienen una altura similar, sugiere poca variación entre las categorías.

__<font color=blue>Casos de uso</font>__

- Comparar la frecuencia o conteo de categorías.
- Mostrar diferencias entre grupos.
- Útil para visualizar datos de encuestas, cifras de ventas o estadísticas categóricas.

__<font color=blue>Documentación</font>__

[Barplot_seaborn](https://seaborn.pydata.org/generated/seaborn.barplot.html)

```python
# Create barplot
sns.barplot(data=data, 
            x="time", 
            y="tip",
            palette=sns.color_palette("tab10", 2))
```

```python
# Crear la figura y los subplots
fig = plt.figure(constrained_layout=True, figsize=(10, 8))
gs = fig.add_gridspec(2, 2)

# Primer plot (Abarcando ambas columnas de la primera fila)
ax1 = fig.add_subplot(gs[0, :])  # [:] que significa?
sns.boxplot(data = data, 
            x = 'day', 
            y = 'total_bill', 
            palette = 'husl', 
            hue="sex",
            ax=ax1) 
ax1.set_title('Boxplot')

# 
ax2 = fig.add_subplot(gs[1, 0])
sns.histplot(data=data, 
             x = 'day',
             palette = 'husl', 
             hue="sex", 
             ax=ax2)
ax2.set_title('Histogram')

# 
ax3 = fig.add_subplot(gs[1, 1])
sns.scatterplot(data=data, 
                x="total_bill", 
                y="tip",
                palette = 'husl', 
                hue='sex',
                ax=ax3)
ax3.set_title('Scatterplot')


# Mostrar el plot
plt.show()
```
