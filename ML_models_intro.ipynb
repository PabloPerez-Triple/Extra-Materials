## 1. Regresión Lineal (Linear Regression)

### **Concepto**
La regresión lineal es un modelo matemático que establece una relación entre una variable dependiente $Y$ y una o más variables independientes $X$. El modelo asume que $Y$ puede predecirse mediante una combinación lineal de $X$, con la fórmula:
$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_n X_n + \epsilon
$$
Donde:
- $ \beta_0 $ es el intercepto.
- $ \beta_1, \beta_2, \dots \beta_n $ son los coeficientes (pendientes) que indican la influencia de cada $X$.
- $ \epsilon $ es el error.

### **Método de Feynman**
Imagina que quieres predecir el precio de una casa usando su tamaño. Piensa en una línea recta que pasa por el medio de todos los datos en un gráfico. Esta línea es tu modelo, y su pendiente y posición te dicen cómo cambian los precios a medida que cambian los tamaños.

### **Referencia gráfica**
Un gráfico típico de regresión lineal muestra una línea ajustada en medio de puntos dispersos:

![Regresión Lineal](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Linear_regression.svg/512px-Linear_regression.svg.png)

### **Utilidad en Data Analysis**
- Predecir tendencias o valores continuos.
- Identificar relaciones entre variables.
- Ejemplo: Predecir ventas futuras basadas en datos históricos.

## 2. Árboles de Decisión (Decision Trees)

### **Concepto**
Un árbol de decisión es un modelo que divide los datos en ramas basándose en decisiones binarias (sí/no). Cada nodo representa una decisión basada en una característica, y cada hoja (final del árbol) contiene una predicción.

### **Método de Feynman**
Piensa en un juego de 20 preguntas: comienzas haciendo preguntas generales y, según las respuestas, refinas tu búsqueda hasta llegar a una solución específica. Cada pregunta en el juego es como un nodo en el árbol.

### **Referencia gráfica**

Un árbol de decisión se ve como un diagrama jerárquico con decisiones en cada nodo, como este:

![Decision Tree](https://upload.wikimedia.org/wikipedia/commons/f/f3/CART_tree_titanic_survivors.png)

Este ejemplo representa un árbol de decisión utilizado para predecir si alguien sobrevivió al desastre del Titanic. Cada nodo indica una decisión basada en una característica (e.g., "¿Sexo masculino?") y cada hoja muestra un resultado (e.g., "Sobrevive: No").

---
### **Utilidad en Data Analysis**
- Clasificación de datos en categorías.
- Identificación de reglas claras para la toma de decisiones.
- Ejemplo: Determinar si un cliente es elegible para un préstamo.

## 3. Bosques Aleatorios (Random Forest)

### **Concepto**
Un bosque aleatorio es un conjunto de muchos árboles de decisión que trabajan juntos. Cada árbol hace una predicción y el modelo final elige la predicción más votada (clasificación) o la media (regresión).

### **Método de Feynman**
Imagina que necesitas decidir qué película ver. Preguntas a 100 amigos y eliges la película que más recomiendan. Cada amigo representa un árbol de decisión en el bosque.

### **Referencia gráfica**



```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Crear la figura
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar árboles (representados como círculos)
tree_positions = [(2, 4), (4, 4), (6, 4), (8, 4)]
for x, y in tree_positions:
    ax.plot(x, y, 'go', markersize=30)  # Árbol como círculo verde
    ax.text(x, y + 0.3, "Tree", ha='center', fontsize=10)

# Dibujar predicciones individuales
predictions = ['A', 'B', 'A', 'A']
for i, (x, y) in enumerate(tree_positions):
    ax.text(x, y - 0.3, f"Predict: {predictions[i]}", ha='center', fontsize=8)

# Conexión de árboles a decisión final
ax.arrow(2, 3.6, 0, -1.2, head_width=0.3, head_length=0.3, fc='black', ec='black')
ax.arrow(4, 3.6, 0, -1.2, head_width=0.3, head_length=0.3, fc='black', ec='black')
ax.arrow(6, 3.6, 0, -1.2, head_width=0.3, head_length=0.3, fc='black', ec='black')
ax.arrow(8, 3.6, 0, -1.2, head_width=0.3, head_length=0.3, fc='black', ec='black')

# Decisión final
ax.plot(5, 1.5, 'ro', markersize=30)  # Nodo final como círculo rojo
ax.text(5, 1.2, "Final Decision: A", ha='center', fontsize=10)

# Ajustar límites y etiquetas
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)
ax.axis('off')  # Sin ejes
plt.title("Random Forest", fontsize=16)
plt.show()
```

## 4. Gradient Boosting

### **Concepto**
Gradient Boosting es un modelo que construye árboles de decisión secuenciales. Cada nuevo árbol corrige los errores del árbol anterior, optimizando continuamente el modelo.

### **Método de Feynman**
Imagina que estás aprendiendo a andar en bicicleta. Al principio te caes, pero aprendes de tus errores y mejoras con cada intento. Gradient Boosting "aprende" de sus errores para mejorar las predicciones.

### **Referencia gráfica**

```python
# Crear la figura
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar árboles (representados como círculos)
tree_positions = [(3, 5), (5, 4), (7, 3)]
for i, (x, y) in enumerate(tree_positions):
    ax.plot(x, y, 'bo', markersize=30)  # Árbol como círculo azul
    ax.text(x, y + 0.3, f"Tree {i+1}", ha='center', fontsize=10)

# Conexión entre árboles
ax.arrow(3, 4.7, 1.7, -0.7, head_width=0.3, head_length=0.3, fc='black', ec='black')
ax.arrow(5, 3.7, 1.7, -0.7, head_width=0.3, head_length=0.3, fc='black', ec='black')

# Decisión final
ax.plot(6, 1.5, 'ro', markersize=30)  # Nodo final como círculo rojo
ax.text(6, 1.2, "Final Prediction", ha='center', fontsize=10)

# Ajustar límites y etiquetas
ax.set_xlim(1, 9)
ax.set_ylim(0, 6)
ax.axis('off')  # Sin ejes
plt.title("Gradient Boosting", fontsize=16)
plt.show()
```




### **Utilidad en Data Analysis**
- Ideal para datos complejos con patrones no lineales.
- Muy usado en competencias de ML por su alta precisión.
- Ejemplo: Predecir precios de casas considerando muchas variables (ubicación, tamaño, antigüedad, etc.).
